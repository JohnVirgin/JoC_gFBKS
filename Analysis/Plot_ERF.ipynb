```python
import numpy as np
import pickle as pk
from matplotlib import pyplot as plt
import scipy.stats as stats
import cartopy
from sklearn.metrics import r2_score as r2_score
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
import matplotlib as mpl
mpl.rc('font',size=16,**{'family':'sans-serif','sans-serif':['Arial']})
```

```python
#read in step response data
path = '/Volumes/eSSD0/Papers/JoC_gFBK/Data/'
pi = pk.load(open(path+'Outputs/f.e12.pi-control.T31_g37.ctl.2_diag_grid.pk','rb'))
x4 = pk.load(open(path+'Outputs/f.e12.abrupt-4xCO2.T31_g37.ctl.2_diag_grid.pk','rb'))
sol = pk.load(open(path+'Outputs/f.e12.abrupt-0.96xTSI.T31_g37.ctl.2_diag_grid.pk','rb'))
```

```python
tas = {}
tas['sol'] = np.nanmean(sol['tas'],axis=(0,1))-np.nanmean(pi['tas'],axis=(0,1))
tas['x4'] = np.nanmean(x4['tas'],axis=(0,1))-np.nanmean(pi['tas'],axis=(0,1))
tas['g1'] = tas['sol']+tas['x4']
```

```python
erf = {}
erf['sol'] = np.nanmean(sol['fnet'],axis=(1))-np.nanmean(pi['fnet'],axis=(1))
erf['x4'] = np.nanmean(x4['fnet'],axis=(1))-np.nanmean(pi['fnet'],axis=(1))
erf['g1'] = erf['sol']+erf['x4']
```

```python
lat = pi['lat']
lon = pi['lon']

y = lat*np.pi/180
coslat = np.cos(y)
coslat = np.tile(coslat[None:,None],(1,lon.size))
```

```python
g1_30yrs = np.nanmean(sol['fnet'],axis=1)+(np.nanmean(x4['fnet'],axis=1))

tval = stats.t.ppf(1-0.05/2,df=58)
ttest = {}
pval = {}

ttest['x4'],pval['x4'] = stats.ttest_ind(np.nanmean(pi['fnet'],axis=1),np.nanmean(x4['fnet'],axis=1), axis=0)
ttest['sol'],pval['sol'] = stats.ttest_ind(np.nanmean(pi['fnet'],axis=1),np.nanmean(sol['fnet'],axis=1), axis=0)
ttest['g1'],pval['g1'] = stats.ttest_ind(np.nanmean(pi['fnet'],axis=1),g1_30yrs,axis=0)


sig = {}
for keys in ttest.keys():
    sig[keys] = (np.abs(ttest[keys]) > tval) & (pval[keys] < 0.05)


binary = {}
for keys in sig.keys():
    binary[keys] = np.zeros([48,96])
    for i in range(48):
        for j in range(96):

            if sig[keys][i,j] == False:
                binary[keys][i,j] = 1
            else:
                pass
```

```python
erf_gam = {}
erf_mn = {}
erf_gse = {}
erf_zm = {}
for keys in erf.keys():
    erf_mn[keys] = np.nanmean(erf[keys],axis=0)
    erf_zm[keys] = np.nanmean(erf[keys],axis=(0,2))
    erf_gam[keys] = np.average(np.nanmean(erf[keys],axis=0),weights=coslat)
```

```python
erf_plot = {}
binary_plot = {}
for keys in erf.keys():
    erf_plot[keys] = np.concatenate((erf_mn[keys][:,48:],erf_mn[keys][:,:48]),axis=1)
    erf_plot[keys],lon_cyc = add_cyclic_point(erf_plot[keys],lon)
    
    binary_plot[keys] = np.concatenate((sig[keys][:,48:],sig[keys][:,:48]),axis=1)
    binary_plot[keys],lon_cyc = add_cyclic_point(binary_plot[keys],lon)
```

```python
color = plt.cm.RdBu_r

bounds = np.arange(-25,25.1,2.5)

fig = plt.figure(figsize=(16,10),dpi=200)
plt.subplot2grid((2, 8), (0, 0), rowspan=1)
plt.title('Zonal Mean')
plt.plot(erf_zm['x4'],lat, color='red')
plt.ylim(-90,90)
plt.xlim(-12,12)
plt.yticks(np.arange(-60,61,30))
plt.xticks(np.arange(-12,12.1,6))
plt.axvline(0,color='k',linestyle='--')
plt.gca().invert_xaxis()

plt.subplot2grid((2, 8), (1, 0), rowspan=1)
plt.title('Zonal Mean')

plt.plot(erf_zm['sol'],lat, color='red')
plt.ylim(-90,90)
plt.xlim(-12,12)
plt.yticks(np.arange(-60,61,30))
plt.xticks(np.arange(-12,12.1,6))
plt.axvline(0,color='k',linestyle='--')
plt.gca().invert_xaxis()


ax0 = plt.subplot2grid((2, 8), (0, 1), colspan=7, rowspan=1,projection=ccrs.PlateCarree())
ax0.set_title(f"a) 4xCO$_{{2(\mathrm{{f}})}}$  [{round(erf_gam['x4'],2)}]",pad=10)
#ax0.set_extent([180, -180, -90, 90], ccrs.PlateCarree())
cont0 = ax0.contourf(lon_cyc,lat,erf_plot['x4'],bounds,cmap=color,extend='both',transform=ccrs.PlateCarree(central_longitude=180))
ax0.contourf(lon_cyc,lat,binary_plot['x4'],colors='none',hatches=['','////'],transform=ccrs.PlateCarree(central_longitude=180))
ax0.coastlines()
g1 = ax0.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')

g1.xlabels_top = False
g1.ylabels_right = False
g1.xlines = False
g1.xlabels_bottom = False

ax1 = plt.subplot2grid((2, 8), (1, 1), colspan=7, rowspan=1,projection=ccrs.PlateCarree())
ax1.set_title(f"b) SOL$_{{\mathrm{{f}}}}$  [{round(erf_gam['sol'],2)}]",pad=10)
#ax0.set_extent([180, -180, -90, 90], ccrs.PlateCarree())
cont1 = ax1.contourf(lon_cyc,lat,erf_plot['sol'],bounds,cmap=color,extend='both',transform=ccrs.PlateCarree(central_longitude=180))
ax1.contourf(lon_cyc,lat,binary_plot['sol'],colors='none',hatches=['','////'],transform=ccrs.PlateCarree(central_longitude=180))
ax1.coastlines()

g2 = ax1.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')

g2.xlabels_top = False
g2.ylabels_right = False
g2.xlabels_bottom = False
g2.xlines = False

plt.subplots_adjust(hspace = 0.4,right=0.8)
cbar_ax = fig.add_axes([0.74, 0.25, 0.02, 0.5])
fig.colorbar(cont0, cax=cbar_ax,drawedges=True,label='Wm$^{-2}$', orientation='vertical')

plt.savefig('ERF_fsst_map.png',bbox_inches='tight')
```
